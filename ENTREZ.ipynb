# PUBMED data

```python
handle = Entrez.einfo(db="pubmed")
record = Entrez.read(handle)
record["DbInfo"]["Description"]  # Description of the PubMed database
```

```python
record["DbInfo"]["Count"]  # Number of articles in PubMed
```

```python
handle = Entrez.esearch(db="pubmed", term="biopython")
record = Entrez.read(handle)
print(record["IdList"])  # List of PubMed IDs for articles matching the search term
```

```python
handle = Entrez.esummary(db="pubmed", id='40160861, 39883659')
records = Entrez.parse(handle)


for record in records:
    print(record['AuthorList'], record['Title'], record['PubDate'])  # List of authors for each article
```

```python
handle = Entrez.efetch(db="pubmed", id="19811691")
print(handle.read())  # Fetch the full text of a specific article
```

# Nucliotide data

```python
handle = Entrez.esearch(db="Nucleotide", retmax = 10, term="Severe Acute Respiratory Syndrome Coronavirus")
record = Entrez.read(handle)
record["IdList"]
```

```python
# Fetching sequences from GenBank
handle = Entrez.efetch(db="nucleotide", id= '3003010001', rettype="gb", retmode="text")
print(handle.read())  # Print the GenBank record for the specified ID
```

```python
handle = Entrez.esearch(db='nucleotide', term='accD[Gene Name] AND "E .coli"[Organism]', retmax="20")
result_list = Entrez.read(handle) # List of IDs for sequences matching the search term
```

```python
id_list = result_list["IdList"]
count = result_list["Count"]  # Total number of sequences matching the search term

print(id_list)
print("\n")
print(count)
```
